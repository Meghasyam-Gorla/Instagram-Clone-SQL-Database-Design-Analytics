```python
from faker import Faker
import csv
import random
```

```python
fake = Faker()
```

```python
with open("users.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["user_id", "username", "email", "full_name", "bio", "profile_pic", "created_at", "is_private", "is_verified"])
    for i in range(1, 7001):
        writer.writerow([f"USR{i:05d}", fake.user_name(), fake.unique.email(), fake.name(),                    
            fake.sentence(nb_words=8), fake.image_url(), fake.date_time_this_decade(), random.choice([0, 1]), random.choice([0, 1])])
```

```python
from datetime import datetime

# 1) Read users and their creation dates
user_created_at = {}

with open("users.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        user_created_at[row["user_id"]] = datetime.fromisoformat(row["created_at"])

user_ids = list(user_created_at.keys())

# 2) Generate posts
with open("posts.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["post_id", "user_id", "caption", "media_url", "created_at", "location"])

    for i in range(1, 10001):
        uid = random.choice(user_ids)
        user_created = user_created_at[uid]
        post_created = fake.date_time_between(start_date=user_created,end_date='now')
        writer.writerow([f"PST{i:06d}", uid, fake.sentence(nb_words=10), fake.image_url(), post_created.isoformat(sep=' '), fake.city()])           
```

```python
# 1) Read post creation dates from posts.csv
post_created_at = {}

with open("posts.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        post_created_at[row["post_id"]] = datetime.fromisoformat(row["created_at"])

post_ids = list(post_created_at.keys())

# 2) Generate comments.csv
with open("comments.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["comment_id", "post_id", "user_id", "comment_text", "created_at"])

    for i in range(1,20001):
        pid = random.choice(post_ids)
        post_created = post_created_at[pid]

        writer.writerow([f"CMT{i:06d}", pid, f"USR{random.randint(1, 5000):05d}", fake.sentence(nb_words=12), 
                         fake.date_time_between(start_date=post_created, end_date='now').isoformat(sep=' ')])
```

```python
# 1) Read post creation dates from posts.csv
post_created_at = {}

with open("posts.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        post_created_at[row["post_id"]] = datetime.fromisoformat(row["created_at"])

post_ids = list(post_created_at.keys())

# 2) Generate likes.csv
with open("likes.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["like_id", "post_id", "user_id", "created_at"])

    for i in range(1, 30001):
        pid = random.choice(post_ids)
        post_created = post_created_at[pid]

        writer.writerow([
            f"LIK{i:06d}", pid, f"USR{random.randint(1, 7001):05d}",# user_id (USR00001 format)
            fake.date_time_between(start_date=post_created, end_date='now').isoformat(sep=' ')])
```

```python
# 1) Read user creation dates from users.csv
user_created_at = {}

with open("users.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        user_created_at[row["user_id"]] = datetime.fromisoformat(row["created_at"])

user_ids = list(user_created_at.keys())

# 2) Generate followers.csv
with open("followers.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["follower_id", "following_id", "followed_at"])

    for _ in range(15000):
        follower = random.choice(user_ids)
        following = random.choice(user_ids)

        # avoid self-follow
        while follower == following:
            following = random.choice(user_ids)
        follower_created = user_created_at[follower]
        followed_at = fake.date_time_between(start_date=follower_created,end_date='now')
        writer.writerow([follower, following, followed_at.isoformat(sep=' ')])
```

```python
hashtags = ["travel", "food", "fitness", "tech", "ai", "cricket", "india", "nature", "music", "coding"]

with open("hashtags.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["hashtag_id", "tag"])

    for i, tag in enumerate(hashtags, start=1):
        writer.writerow([f"HSH{i:03d}", tag])
```

```python
# Read post_ids
post_ids = []
with open("posts.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        post_ids.append(row["post_id"])

# Read hashtag_ids
hashtag_ids = []
with open("hashtags.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        hashtag_ids.append(row["hashtag_id"])

# Generate post_hashtags.csv
with open("post_hashtags.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["post_id", "hashtag_id"])

    for _ in range(20000):
        writer.writerow([random.choice(post_ids), random.choice(hashtag_ids)])
```
